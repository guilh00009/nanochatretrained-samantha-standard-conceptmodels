```python
from huggingface_hub import hf_hub_download
from pathlib import Path
import shutil

repo_id = "Guilherme34/nanochat-retrained-pytorch-duplicated"
files = [
    "model_000700.pt",
    "meta_000700.json",
    "token_bytes.pt",
    "tokenizer.pkl",
]

# Output directory
out_dir = Path("/content/nanochat-inference/model")
out_dir.mkdir(parents=True, exist_ok=True)

downloaded = []
for fname in files:
    cached_path = hf_hub_download(
        repo_id=repo_id,
        filename=fname,
        revision="main"
    )
    target = out_dir / fname
    shutil.copy2(cached_path, target)
    downloaded.append(str(target))

print("✅ Saved files to:", out_dir)
print("\n".join(downloaded))
```
